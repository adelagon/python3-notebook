# Welcome to Python 3 Training

## How to setup this notebook

This guide assumes that you are using MacOS Sierra v10.12.x.

* Install **pyenv**

  ```bash
  brew install pyenv
  ```

* Append this to your **~/.bash_profile**

  ```bash
  eval "$(pyenv init -)"
  ```

* Install python 3.7.0 using pyenv

  ```bash
  pyenv install 3.7.0
  ```

* Set your python version to 3.7.0

  ```bash
  pyenv global 3.7.0
  ```

* Confirm your python version

  ```
  $ python
  Python 3.7.0 (default, Sep 24 2018, 15:04:24)
  [Clang 8.1.0 (clang-802.0.42)] on darwin
  Type "help", "copyright", "credits" or "license" for more information.
  ```
* Clone the python-training-notebook

  ```bash
  $ git clone git@code.corp.voyager.ph:voyager-commons/python-training-notebook.git
  $ cd python-training-notebook.git
  ```
* Install Dependencies

  ```bash
  $ pip install -r requirements.txt
  ```
* Install jupyter extensions

  ```bash
  $ ./install-nbextensions.sh
  ```
* Run jupyter

  ```bash
  $ jupyter notebook
  ```
  
  A browser window should open on http://localhost:8888
  
**[IMPORTANT NOTE]** These steps should only be ran once. If you want to run the notebook again just do

  ```bash
  $ jupyter notebook
  ```
